```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Ensure datetime conversions
sqf_data['STOP_FRISK_DATE'] = pd.to_datetime(sqf_data['STOP_FRISK_DATE'])
sqf_data['STOP_FRISK_TIME'] = pd.to_datetime(sqf_data['STOP_FRISK_TIME'], format='%H:%M:%S').dt.time


```

```python
# Group by Month and calculate mean for binary columns
month_data = sqf_data.groupby('MONTH2')[['SUSPECT_ARRESTED_FLAG', 'FRISKED_FLAG', 'SEARCHED_FLAG']].mean() * 100

# Calculate unjustified rates
month_data['Unjustified Stop Rate'] = 100 - month_data['SUSPECT_ARRESTED_FLAG']
month_data['Unjustified Frisk Rate'] = month_data['FRISKED_FLAG'] * (1 - month_data['SUSPECT_ARRESTED_FLAG'] / 100)
month_data['Unjustified Search Rate'] = month_data['SEARCHED_FLAG'] * (1 - month_data['SUSPECT_ARRESTED_FLAG'] / 100)

# Ensure proper month ordering
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
               'September', 'October', 'November', 'December']
month_data = month_data.reindex(month_order)

# Plotting the bar chart for monthly outcomes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot unjustified rates as grouped bars
month_data[['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate']].plot(
    kind='bar', ax=ax, width=0.75, legend=False)


plt.title('Monthly Disparities in Unjustified Stop, Frisk, and Search Rates')
plt.xlabel('Month')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.legend(['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate'],
           title='Outcome Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# Group by Day of the Week and calculate mean for binary columns
day_data = sqf_data.groupby('DAY2')[['SUSPECT_ARRESTED_FLAG', 'FRISKED_FLAG', 'SEARCHED_FLAG']].mean() * 100

# Calculate unjustified rates
day_data['Unjustified Stop Rate'] = 100 - day_data['SUSPECT_ARRESTED_FLAG']
day_data['Unjustified Frisk Rate'] = day_data['FRISKED_FLAG'] * (1 - day_data['SUSPECT_ARRESTED_FLAG'] / 100)
day_data['Unjustified Search Rate'] = day_data['SEARCHED_FLAG'] * (1 - day_data['SUSPECT_ARRESTED_FLAG'] / 100)

# Ensure proper weekday ordering
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_data = day_data.reindex(day_order)

# Plotting the bar chart for day of the week outcomes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot unjustified rates as grouped bars
day_data[['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate']].plot(
    kind='bar', ax=ax, width=0.75, legend=False)

# Customize plot
plt.title('Day of the Week Disparities in Unjustified Stop, Frisk, and Search Rates')
plt.xlabel('Day of the Week')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Place legend outside the plot
plt.legend(['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate'],
           title='Outcome Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

## Impact of Demeanor on Police Suspicion

```python
# Select specific demeanors to include in the analysis
selected_demeanors = [
    'CALM AND COOPERATIVE',
    'COMPLIANT',
    'NERVOUS',
    'AGGRESSIVE',
    'AGITATED',
    'UNCOOPERATIVE',
    'UPSET',
    'SCARED'
]

# Filter the dataset for the selected demeanors
filtered_demeanor_data = sqf_data[sqf_data['DEMEANOR_OF_PERSON_STOPPED'].isin(selected_demeanors)]

# Group by the selected demeanors and calculate percentages
demeanor_outcomes = filtered_demeanor_data.groupby('DEMEANOR_OF_PERSON_STOPPED')[binary_columns].mean() * 100

# Calculate unjustified rates
demeanor_outcomes['Unjustified Stop Rate'] = 100 - demeanor_outcomes['SUSPECT_ARRESTED_FLAG']
demeanor_outcomes['Unjustified Frisk Rate'] = demeanor_outcomes['FRISKED_FLAG'] * (1 - demeanor_outcomes['SUSPECT_ARRESTED_FLAG'] / 100)
demeanor_outcomes['Unjustified Search Rate'] = demeanor_outcomes['SEARCHED_FLAG'] * (1 - demeanor_outcomes['SUSPECT_ARRESTED_FLAG'] / 100)

# Plotting the bar chart for the selected demeanors
fig, ax = plt.subplots(figsize=(14, 8))


demeanor_outcomes[['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate']].plot(kind='bar', ax=ax, width=0.75, legend=False)
plt.title('Demeanor-Based Disparities in Unjustified Stop, Frisk, and Search Rates (Selected)')
plt.xlabel('Demeanor of Person Stopped')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=70)


plt.legend(['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate'],
           title='Outcome Type', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.subplots_adjust(left=0.15, right=0.8, top=0.9, bottom=0.3)

plt.show()
```

The data suggests that hostile, anxious, or aggressive demeanors correlate with a higher likelihood of justified stops, frisks, and searches, as indicated by the lower unjustified rates for these actions. Conversely, individuals who appear calm or cooperative are more likely to experience stops, frisks, and searches that do not lead to arrests, which may indicate a lack of sufficient justification in these cases. This pattern implies that demeanor plays a role in the perceived need for stops and searches, with more confrontational behaviors aligning more closely with arrest outcomes.

## Cyclical Encoding of Temporal Features

```python


# Drop the MONTH2 column
sqf_model_data = sqf_model_data.drop(columns=['MONTH2'])

# Extract day of the month and month (numerical format) from STOP_FRISK_DATE
sqf_model_data['Day_of_Month'] = sqf_model_data['STOP_FRISK_DATE'].dt.day
sqf_model_data['Month'] = sqf_model_data['STOP_FRISK_DATE'].dt.month

# Convert DAY2 (day of the week) from string to numerical format (e.g., Monday = 0, Sunday = 6)
sqf_model_data['Day_of_Week'] = pd.to_datetime(sqf_model_data['STOP_FRISK_DATE']).dt.dayofweek

# Apply cyclical encoding for Hour, Day of the Week, and Month
# Encode Hour as a cyclical feature
sqf_model_data['Hour_sin'] = np.sin(2 * np.pi * sqf_model_data['Hour'] / 24)
sqf_model_data['Hour_cos'] = np.cos(2 * np.pi * sqf_model_data['Hour'] / 24)

# Encode Day of the Week as a cyclical feature
sqf_model_data['Day_of_Week_sin'] = np.sin(2 * np.pi * sqf_model_data['Day_of_Week'] / 7)
sqf_model_data['Day_of_Week_cos'] = np.cos(2 * np.pi * sqf_model_data['Day_of_Week'] / 7)

# Encode Month as a cyclical feature
sqf_model_data['Month_sin'] = np.sin(2 * np.pi * sqf_model_data['Month'] / 12)
sqf_model_data['Month_cos'] = np.cos(2 * np.pi * sqf_model_data['Month'] / 12)

# Drop old columns since they are no longer needed
sqf_model_data = sqf_model_data.drop(columns=['STOP_FRISK_DATE','STOP_FRISK_TIME', 'DAY2', 'Hour', 'Day_of_Week', 'Month'])
```

## Target Feature Engineering: Unjustified Stop, Frisk, Search

```python
sqf_model_data['Unjustified_Stop'] = (sqf_model_data['SUSPECT_ARRESTED_FLAG'] == 0).astype(int)
sqf_model_data['Unjustified_Frisk'] = ((sqf_model_data['FRISKED_FLAG'] == 1) & (sqf_model_data['SUSPECT_ARRESTED_FLAG'] == 0)).astype(int)
sqf_model_data['Unjustified_Search'] = ((sqf_model_data['SEARCHED_FLAG'] == 1) & (sqf_model_data['SUSPECT_ARRESTED_FLAG'] == 0)).astype(int)
```

```python
sqf_model_data = sqf_model_data.drop(columns=['SUSPECTED_CRIME_DESCRIPTION', 'SUSPECT_ARREST_OFFENSE','SUSPECT_ARRESTED_FLAG','FRISKED_FLAG','SEARCHED_FLAG'], axis=1)
```

### Correlation Matrix & Factors

### VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Convert boolean columns (True/False) to integer (1/0)
sqf_model_data_vif = sqf_model_data_dummified.copy()  # Make a copy to avoid altering the original data
sqf_model_data_vif = sqf_model_data_vif.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)

features = sqf_model_data_vif.select_dtypes(include=['float64', 'int64', 'int32'])

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = features.columns
vif_data['VIF'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
```

```python
high_vif = vif_data[vif_data['VIF'] > 10]


print(high_vif)
```
